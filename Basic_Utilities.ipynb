# Rename in Order

```python
import os
from natsort import natsorted 
```

```python
def rename(path, new_name, extension):
  os.chdir(path)

  for(i, filename) in enumerate(natsorted(os.listdir(path))):
    os.rename(src=filename, dst='{}{}{}'.format(new_name, i, extension))
```

```python
#rename("PATH","NEW NAME","EXTENSION (.ext)")
rename("/content/drive/MyDrive/Expensive Cars/BGR2GRAY/Images","exp_grey_",".jpg")
```
